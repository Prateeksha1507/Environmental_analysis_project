```python
# Import required libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define pollutants
pollutants = ['PM2.5 (µg/m³)','PM10 (µg/m³)','NO (µg/m³)','NO2 (µg/m³)',
              'CO (mg/m³)','SO2 (µg/m³)','Ozone (µg/m³)']

# Define distributions to test
distributions = {
    'Lognormal': stats.lognorm,
    'Gamma': stats.gamma,
    'Weibull': stats.weibull_min
}

# Load pre- and post-SEZ data (replace with your file paths)
pre_df = pd.read_csv('data/sector51gurugram2021.csv')
post_df = pd.read_csv('data/sector51gurugram2023.csv')

# Optional: Drop rows with NaNs
pre_df = pre_df.dropna(subset=pollutants)
post_df = post_df.dropna(subset=pollutants)

# Function to fit distributions and calculate AIC
def fit_distributions(data, pollutant):
    results = {}
    values = data[pollutant].values
    for name, dist in distributions.items():
        try:
            params = dist.fit(values)
            log_likelihood = np.sum(dist.logpdf(values, *params))
            k = len(params)
            aic = 2 * k - 2 * log_likelihood
            results[name] = {'params': params, 'aic': aic}
        except Exception as e:
            results[name] = {'params': None, 'aic': np.inf}
    return results

# Function to perform Chi-square test
def chi_square_test(data, pollutant, dist_name, params):
    values = data[pollutant].values
    hist, bin_edges = np.histogram(values, bins='auto')
    if dist_name == 'Lognormal':
        cdf_vals = stats.lognorm.cdf(bin_edges, *params)
    elif dist_name == 'Gamma':
        cdf_vals = stats.gamma.cdf(bin_edges, *params)
    elif dist_name == 'Weibull':
        cdf_vals = stats.weibull_min.cdf(bin_edges, *params)
    expected = len(values) * np.diff(cdf_vals)
    expected = np.where(expected < 0.001, 0.001, expected)
    chi2 = np.sum((hist - expected) ** 2 / expected)
    df = len(hist) - len(params) - 1
    p = 1 - stats.chi2.cdf(chi2, df)
    return {'chi2': chi2, 'p_value': p, 'df': df}

# Analyze and store results
results = {}

for period, df in [('pre', pre_df), ('post', post_df)]:
    for pollutant in pollutants:
        fit_result = fit_distributions(df, pollutant)
        best_dist = min(fit_result, key=lambda k: fit_result[k]['aic'])
        best_params = fit_result[best_dist]['params']
        chi_result = chi_square_test(df, pollutant, best_dist, best_params)
        results[f"{period}_{pollutant}"] = {
            'best_fit': best_dist,
            'chi2': chi_result['chi2'],
            'p_value': chi_result['p_value']
        }

# Display results
for key, val in results.items():
    print(f"{key}: Best fit = {val['best_fit']}, Chi² = {val['chi2']:.2f}, p = {val['p_value']:.4f}")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_fit_test(data, dist_names, bins=20):
    data = data.dropna()
    results = {}

    # Histogram binning (same for all distributions)
    observed_freq, bin_edges = np.histogram(data, bins=bins)
    total = len(data)

    for dist_name in dist_names:
        dist = getattr(stats, dist_name)
        params = dist.fit(data)

        # CDF-based expected frequencies
        cdf_values = dist.cdf(bin_edges, *params)
        expected_freq = total * np.diff(cdf_values)

        # Avoid division by zero
        expected_freq = np.where(expected_freq == 0, 1e-5, expected_freq)

        # Chi-square statistic
        chi2_stat = np.sum((observed_freq - expected_freq) ** 2 / expected_freq)
        dof = bins - len(params) - 1
        p_val = 1 - stats.chi2.cdf(chi2_stat, df=dof)

        results[dist_name] = {
            'chi2': chi2_stat,
            'p_value': p_val,
            'dof': dof,
            'params': params
        }

    return results

# Choose pollutant
pollutant = 'PM2.5 (µg/m³)'
dist_names = ['gamma', 'lognorm', 'weibull_min']  # note: Weibull is weibull_min in scipy

# Run for Pre and Post SEZ
pre_results = chi_square_fit_test(df_pre[pollutant], dist_names)
post_results = chi_square_fit_test(df_post[pollutant], dist_names)

# Display results
def display_fit_results(title, results):
    print(f"\n--- {title} ---")
    for dist, stats_dict in results.items():
        print(f"{dist.capitalize():<12} | Chi² = {stats_dict['chi2']:.2f} | p = {stats_dict['p_value']:.4f} | DOF = {stats_dict['dof']}")

display_fit_results("Pre-SEZ PM2.5 Distribution Fit", pre_results)
display_fit_results("Post-SEZ PM2.5 Distribution Fit", post_results)
```
